```r
# Has estimates close to a:
# BetaPrime model
# InvGamma model
# InverseWeibull model
# Loggamma model
# Loglogistic model
# Lognormal model
# Lomax model
# Pareto model
```

```r
tests <- function(lis){ 
    print(mlbetapr(lis))
    cat("\n")
    print(mlcauchy(lis))
    cat("\n")
    print(mlexp(lis))
    cat("\n")
    print(mlgamma(lis))
    cat("\n")
    print(mlgumbel(lis))
    cat("\n")
    print(mlinvgamma(lis))
    cat("\n")
    print(mlinvgauss(lis))
    cat("\n")
    print(mlinvweibull(lis))
    cat("\n")
    print(mllaplace(lis))
    cat("\n")
    print(mllgamma(lis))
    cat("\n")
    print(mlllogis(lis))
    cat("\n")
    print(mllnorm(lis))
    cat("\n")
    print(mllogis(lis))
    cat("\n")
    print(mllomax(lis))
    cat("\n")
    print(mlnaka(lis))
    cat("\n")
    print(mlnorm(lis))
    cat("\n")
    print(mlpareto(lis))
    cat("\n")
    print(mlpower(lis))
    cat("\n")
    print(mlrayleigh(lis))
    cat("\n")
    print(mlunif(lis))
    cat("\n")
    print(mlweibull(lis))
    cat("\n")
}
print('sum')
tests(df$sum)
```
