```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
%matplotlib inline
```

```python
samples = np.loadtxt("samples.txt")
samples.shape
```

```python
mid = int(samples.shape[1]/2)
for k in range(101):
    
    plt.rc('figure', figsize=(5, 5))
    plt.xlim([-6,6])
    plt.ylim([0,0.50])
    plt.xticks(np.arange(-6, 7, step=1))
    
    tmp = None
    if k != 0:
        tmp = samples[:,mid-k] - samples[:,mid+k]
    else:
        tmp = samples[:,mid]
    tmp = (tmp-np.mean(tmp)) / np.std(tmp)
    
    increments=0.10
    plt.hist(tmp, density=True, bins=np.arange(min(tmp), max(tmp) + increments, increments))
    
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    mu, std = norm.fit(tmp)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=0.8)
    
    t = 'k = ' + str(k)
    plt.title(t)
    plt.show()
```
