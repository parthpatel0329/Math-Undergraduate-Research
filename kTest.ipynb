```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
%matplotlib inline
```

```python
samples = np.loadtxt("samples.txt")
samples.shape
```

```python
mid = int(samples.shape[1]/2)
for k in range(101):
    tmp = None
    if k != 0:
        tmp = samples[:,mid-k] - samples[:,mid+k]
    else:
        tmp = samples[:,mid]
    tmp = (tmp-np.mean(tmp)) / np.std(tmp)
    
    plt.hist(tmp, density=True)
    
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    mu, std = norm.fit(tmp)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2)
    
    t = 'k = ' + str(k)
    plt.title(t)
    plt.show()
```
